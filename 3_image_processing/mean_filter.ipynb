<h1>Mean Filter</h1>

Example 3.2

```python
# import python packages
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import convolve
from skimage.data import camera
```

Define input image

```python
image = camera()
```

Definition of the mean filter as in Def. 3.1

```python
def mean_filter_kernel(n):
    kernel = (1/(n**2)) * np.ones((n, n))
    return kernel
```

Convolve input by the filter kernel.

```python
mean1 = convolve(image, mean_filter_kernel(3))
mean2 = convolve(image, mean_filter_kernel(5))
mean3 = convolve(image, mean_filter_kernel(11))
```

Plot the results

```python
fig, axes = plt.subplots(ncols=4, figsize=(16, 4))

axes[0].imshow(image, cmap=plt.cm.gray)
axes[0].set_title('Original')

axes[1].imshow(mean1, cmap=plt.cm.gray)
axes[1].set_title('I*K_3')

axes[2].imshow(mean2, cmap=plt.cm.gray)
axes[2].set_title('I*K_5')

axes[3].imshow(mean3, cmap=plt.cm.gray)
axes[3].set_title('I*K_11')

for ax in axes:
    ax.axis('off')

plt.tight_layout()
plt.savefig('mean_filter', bbox_inches='tight')
plt.show()
```
