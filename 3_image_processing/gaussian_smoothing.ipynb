<h1>Gaussian smoothing</h1>

Example 3.5

```python
# import python packages
import numpy as np
import matplotlib.pyplot as plt
import skimage.filters
from skimage.data import camera
```

Filter image by different kernels.

```python
image = camera()
smoothened1 = skimage.filters.gaussian(image)
smoothened2 = skimage.filters.gaussian(image, sigma=4)
smoothened3 = skimage.filters.gaussian(image, sigma=20)
```

Plot results.

```python
fig, axes = plt.subplots(ncols=4, figsize=(16, 4))

axes[0].imshow(image, cmap=plt.cm.gray)
axes[0].set_title('Original')

axes[1].imshow(smoothened1, cmap=plt.cm.gray)
axes[1].set_title('Sigma=1')

axes[2].imshow(smoothened2, cmap=plt.cm.gray)
axes[2].set_title('Sigma=4')

axes[3].imshow(smoothened3, cmap=plt.cm.gray)
axes[3].set_title('Sigma=20')

for ax in axes:
    ax.axis('off')

plt.tight_layout()
plt.savefig('gaussian_smoothing', bbox_inches='tight')
plt.show()
```
