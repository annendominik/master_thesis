<h1>Resampling</h1>

<h2>Downsampling</h2>

Figure 3.8

```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
from skimage.data import camera
from skimage.transform import downscale_local_mean, resize
```

Downsample image

```python
image = camera()
downsampled2 = downscale_local_mean(image, (2, 2))
downsampled10 = downscale_local_mean(image, (10, 10))
```

Plot results

```python
fig, axes = plt.subplots(ncols=3, figsize=(12, 4))

axes[0].imshow(image, cmap=plt.cm.gray)
axes[0].set_title('Original')

axes[1].imshow(downsampled2, cmap=plt.cm.gray)
axes[1].set_title('Downsampled by 2')

axes[2].imshow(downsampled10, cmap=plt.cm.gray)
axes[2].set_title('Downsampled by 10')


plt.tight_layout()
plt.savefig('downsample')
plt.show()
```

<h2>Upsampling</h2>

Figure 3.9

Perform upsampling by bilinear interpolation.

```python
upsampled2 = resize(downsampled2, camera().shape, order=1)
```

Plot results

```python
fig, axes = plt.subplots(ncols=3, figsize=(12, 4))

axes[0].imshow(image, cmap=plt.cm.gray)
axes[0].set_title('Original')

axes[1].imshow(downsampled2, cmap=plt.cm.gray)
axes[1].set_title('Downsampled by 2')

axes[2].imshow(upsampled2, cmap=plt.cm.gray)
axes[2].set_title('Upsampled to original size')


plt.tight_layout()
plt.savefig('upsample')
plt.show()
```

Compute average error in grayscale value.

```python
error = camera() - upsampled2
average_error = np.mean(error)
print('The average error of any image pixel is ' + str(average_error))
```
