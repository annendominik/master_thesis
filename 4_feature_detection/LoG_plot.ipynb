<h1> Laplacian of Gaussian curve</h1>

Figure 4.5

```python
# import python packages
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
```

Define LoG function.

```python
x = np.linspace(-4, 4, 100)
x, y = np.meshgrid(x, x)
sigma = 1

LoG = - 1 / (np.pi * sigma**4) * (1 - (x**2 + y**2)/(2 * sigma**2)) * np.exp(-(x**2 + y**2)/(2 * sigma**2))
```

Create 3d plot.

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('LoG(x,y)')
plt.tight_layout()
ax.plot_surface(x, y, LoG, cmap=cm.gray)
plt.savefig('LoG')
plt.show()
```
