Example 4.7

```python
# import python modules
import numpy as np
import matplotlib.pyplot as plt

from skimage import filters
from skimage.data import camera, horse
from skimage.util import compare_images
```

Apply Sobel to two images and display the result.

```python
fig, axes = plt.subplots(ncols=2, sharex=True, sharey=True,
                         figsize=(8, 4))

axes[0].imshow(camera(), cmap=plt.cm.gray)

%timeit filters.sobel(camera())
filtered_camera = filters.sobel(camera())

axes[1].imshow(filtered_camera, cmap=plt.cm.gray)

for ax in axes:
    ax.axis('off')

plt.tight_layout()
plt.savefig('sobel_camera')
plt.show()
```

```python
fig, axes = plt.subplots(ncols=2, sharex=True, sharey=True,
                         figsize=(8, 4))

axes[0].imshow(horse(), cmap=plt.cm.gray)

%timeit filters.sobel(horse())
filtered_horse = filters.sobel(horse())

axes[1].imshow(filtered_horse, cmap=plt.cm.gray)

for ax in axes:
    ax.axis('off')

plt.tight_layout()
plt.savefig('sobel_horse')
plt.show()
```
