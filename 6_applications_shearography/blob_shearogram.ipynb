<h1> Blob detection on a shearogram </h1>

```python
from math import sqrt
from skimage.feature import blob_doh
from skimage.color import rgb2gray, rgba2rgb

import matplotlib.pyplot as plt
```

Figure 6.7

Apply DoH to shearogram

```python
image = plt.imread('shearogramm.png')
image = rgba2rgb(image)
image = rgb2gray(image)

blobs_doh = blob_doh(image, max_sigma=31, threshold=.004)

fig, ax = plt.subplots()

plt.title('Prototype Shearogram')
plt.imshow(image, cmap=plt.cm.gray)
for blob in blobs_doh:
    y, x, r = blob
    c = plt.Circle((x, y), r, color='red', linewidth=2, fill=False)
    ax.add_patch(c)
plt.axis('off')

plt.tight_layout()
plt.savefig('blobs_shear')
plt.show()
```

Figure 6.8

Extract blobs at large scale

```python
image = plt.imread('shearogramm.png')
image = rgba2rgb(image)
image = rgb2gray(image)

blobs_doh = blob_doh(image, max_sigma=31, threshold=.004)

fig, ax = plt.subplots()

thres = 30

plt.title('Prototype Shearogram')
plt.imshow(image, cmap=plt.cm.gray)
for blob in blobs_doh:
    y, x, r = blob
    if r > thres:
        c = plt.Circle((x, y), r, color='red', linewidth=2, fill=False)
        ax.add_patch(c)
plt.axis('off')

plt.tight_layout()
plt.savefig('main_blobs_shear')
plt.show()
```

Figure 6.7

Apply DoH to denoised shearogram

```python
image2 = plt.imread('denoised.png')
image2 = rgba2rgb(image2)
image2 = rgb2gray(image2)


blobs_doh = blob_doh(image2, max_sigma=40, threshold=.004)

fig, ax = plt.subplots()

plt.title('Filtered Shearogram')
plt.imshow(image2, cmap=plt.cm.gray)
for blob in blobs_doh:
    y, x, r = blob
    c = plt.Circle((x, y), r, color='red', linewidth=2, fill=False)
    ax.add_patch(c)
plt.axis('off')

plt.tight_layout()
plt.savefig('blobs_shear2')
plt.show()
```

Figure 6.8

Extract blobs from largest scales.

```python
blobs_doh = blob_doh(image2, max_sigma=40, threshold=.005)

fig, ax = plt.subplots()

thres = 35

plt.title('Filtered Shearogram')
plt.imshow(image2, cmap=plt.cm.gray)
for blob in blobs_doh:
    y, x, r = blob
    if r > thres:
        c = plt.Circle((x, y), r, color='red', linewidth=2, fill=False)
        ax.add_patch(c)
plt.axis('off')

plt.tight_layout()
plt.savefig('main_blobs_shear2')
plt.show()
```
